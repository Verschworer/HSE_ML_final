Обрезаем до 2  последних слов

```python
for x in ing_dict_train:
    for k in x:
        z=x.get(k)
        x.update({k:[' '.join (i.split(' ')[-2:]) for i in z]})
```

```python
import pattern
from pattern.en import lemma

for x in ing_dict_train:
    for k in x:
        z=x.get(k)
        x.update({k:[' '.join (lemma(i).split(' ')) for i in z]})
```

2 слова + лемма

```python
for x in ing_dict_train:
    for k in x:
        z=x.get(k)
        x.update({k:[' '.join (lemma(i).split(' ')[-2:]) for i in z]})
```

```python
ing_dict_train
```

```python
from sklearn.feature_extraction import DictVectorizer
vectorizer = DictVectorizer()
```

```python
train = vectorizer.fit_transform(ing_dict_train)
```

```python
train_df = pd.DataFrame(train.toarray(), index=df.index,columns=vectorizer.feature_names_)
train_df
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression(multi_class='ovr', max_iter=2000, random_state=10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_df, y, test_size=6000, random_state=10)
```

```python
model.fit(X_train, y_train)
```

```python
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
```

```python
print(accuracy_score(y_train, pred_train))
print(accuracy_score(y_test, pred_test))
```

```python
cross_val_score(model,train_df, y, cv=5, scoring='accuracy',n_jobs=-1).mean()
```

submission

```python
test_df=pd.read_json('test.json').set_index('id')

ing_dict_test = test_df[['ingredients']].to_dict(orient='records')

for x in ing_dict_test:
    for k in x:
        z=x.get(k)
        x.update({k:[' '.join (lemma(i).split(' ')[-2:]) for i in z]})
        
test = vectorizer.fit_transform(ing_dict_test)

test = pd.DataFrame(test.toarray(), index=test_df.index)
```
